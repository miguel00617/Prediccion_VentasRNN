# Predicción de Ventas (LSTM) #

## Redes Neuronales con LSTM(Series de Tiempo) ##

Este modelo de Machine Learning, no se compone de una simple neurona o una pequeña red neuronal, este modelo contiene redes LSTM(Long Short-Term Memory), tiene en cuenta series de tiempo de datos históricos, donde predice, no simples filas, sino secuencias de tiempo. Ideal para negocios que se dediquen a la venta de productos(Ropa, Víveres, herramientas, Supermercados, Etc) y quieran conocer que tan bien irá la venta.

       6to: Conclusión

## 2do: Preparación para el Modelo

### Nos cerraremos a solo una tienda y un solo Departamento, para facilitar al LSTM, ya que al tener varias fechas cruzadas por las tiendas y departamentos, esto mezclaría al LSTM con varias series de tiempo, por ello mejor la simplifacremos por Tienda 1 y Depto 1

```python
# Por el momento dejaremos de lado IsHoliday, para centrarnos en la fecha.
train_filtrado = train[(train['Store'] == 1) & (train['Dept'] == 1)].copy()
```

```python
train_limpio = train_filtrado[['Date', 'Weekly_Sales']].sort_values('Date')
```

```python
train_limpio.head(10)
```

#### Ya teniendo limpio nustra variable de train, veamos su gráfica

```python
plt.figure(figsize=(14,5))
plt.plot(train_limpio['Date'], train_limpio['Weekly_Sales'])
plt.title(f"Ventas semanales")
plt.xlabel("Fecha")

plt.ylabel("Ventas")
plt.show()
```

#### Podemos observar los picos, la estacionalidad y la tendencia de las ventas entre las fechas.

### El siguiente paso es el separar los datos que tenemos para entrenar y validar

```python
ventas = train_limpio['Weekly_Sales'].values.reshape(-1, 1)
```

```python
#Creamos al escalador
scaler = MinMaxScaler(feature_range=(0, 1))
```

```python
train_tamaño = int(len(ventas) * 0.8)
train_data = ventas[:train_tamaño]
validation_data = ventas[train_tamaño:]
```

#### Escalemos el Train para adaptarlo al modelo y reducir el impacto de llos picos


```python
train_scaled = scaler.fit_transform(train_data)
```

```python
validation_scaled = scaler.transform(validation_data)
```

#### Crearemos ventanas para dividir las series temporales

```python
Ventana = 52
#Ventana de 12 semanas, o sea, 3 meses
```

```python
#prueba
context = train_scaled[-Ventana:]
validation_with_context = np.concatenate([context, validation_scaled])
```

```python
#Creamos la función que nos ayudará a crear estas secuencias
def create_sequences(data,window):
    X, y = [], []
    for i in range(len(data) - window):
        X.append(data[i:i+window])
        y.append(data[i+window])
    return np.array(X), np.array(y)
```

#### Aplicaremos esta función con nuestros datos


```python
X_train, y_train = create_sequences(train_scaled, Ventana)
X_val, y_val     = create_sequences(validation_with_context, Ventana)
```

```python
X_train.shape, y_train.shape
```

### Teniendo listas las variable spara entrenar al modelo, empezemos a crearlo

## 3ro: Creación del Modelo

```python
model = Sequential()
```

```python
model.add(LSTM(128,return_sequences=True,input_shape=(X_train.shape[1], X_train.shape[2])))
model.add(LSTM(64))
model.add(Dropout(0.2))
model.add(Dense(1))
```

```python
# Compilador del modelo
model.compile(optimizer='adam',loss='mse',metrics=['mae'])
```

```python
#Le aplicaremos EarlyStopping para que el modelo no se sobreentrene, sino que pare de entrenarse cuanod alcanzó su máximo punto
```

```python
early_stop = EarlyStopping(monitor='val_loss',patience=40,restore_best_weights=True)
```

## 4to: Entrenamiento del Modelo

##### Entrenamos al modelo

```python
history = model.fit(
    X_train,
    y_train,
    validation_data=(X_val, y_val),
    epochs=100,
    batch_size=16,
    shuffle=False,
    callbacks=[early_stop],
    verbose=1
)
```

#### Veamos que tal le fue al modelo:

```python
plt.figure(figsize=(10,4))
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.legend()
plt.title("Pérdida durante el entrenamiento")
plt.show()
```

#### El modelo presenta que validation, esta por debajo de train, pero sigue sus cambios, esto presenta que quizás train tenia datos mas volatiles, o que por la regularización con dropout y early stopping, no memorice los datos, o por la escala que cambiamos. 

#### El modleo no presenta overfitting

### Validemos el Modelo

```python
val_predictions = model.predict(X_val)
```

#### Invertimos la escala de la predicción del modelo

```python
val_predictions_inv = scaler.inverse_transform(val_predictions)
y_val_inv = scaler.inverse_transform(y_val.reshape(-1, 1))
```

#### Establecemos el RMSE y MAE

```python
mae = mean_absolute_error(y_val_inv, val_predictions_inv)
rmse = np.sqrt(mean_squared_error(y_val_inv, val_predictions_inv))
```

```python
mae, rmse
```

##### MAE: El error en predicción
##### RMSE: Cuanto castigó al modelo para que sea exacto

```python
plt.figure(figsize=(14,5))
plt.plot(y_val_inv, label='Ventas reales')
plt.plot(val_predictions_inv, label='Ventas predichas')
plt.legend()
plt.title("Ventas reales vs predichas (Validation)")
plt.show()
```

## 5to: Evaluación del Modelo

#### Importemos Test.csv

```python
#Al no tener ventas semanales(Weekly_Sales), vamos a comparar con la ultima semana de ventas de train
ultima_ventana = scaler.transform(ventas[-Ventana:])
ultima_ventana = ultima_ventana.reshape(1, Ventana, 1)
```

```python
#Empezaremos a crear un bucle para la predicción teniendo en cuenta la ultima semana de train y las fechas posteriores de test
```

```python
n_forecast = 39
#forecast solamente de 39 semanas
```

```python
#Defionimos variables
predicciones = []
ventana_actual = ultima_ventana.copy()
```

```python
print("n_forecast =", n_forecast)
```

```python
#bucle for para predecir cada semana, en base a las ventas pasadas
for i in range(n_forecast):
    pred = model.predict(ventana_actual, verbose=0)
    pred = np.clip(pred, 0, 1)
    predicciones.append(pred[0,0])
    ventana_actual = np.append(
    ventana_actual[:,1:,:],
    pred.reshape(1,1,1),
    axis=1)
```

```python
#Reinvertimos la escala: 
predicciones = np.array(predicciones).reshape(-1, 1)
predicciones_real = scaler.inverse_transform(predicciones)
```

```python
#Ventas predichas
ultima_fecha = train_limpio['Date'].max()

fechas_futuras = pd.date_range(
    start=ultima_fecha + pd.Timedelta(weeks=1),
    periods=n_forecast,
    freq='W-FRI'
)
```

```python
fechas_futuras = pd.DataFrame({'Date': fechas_futuras})
```

```python
print("Cantidad fechas:", len(fechas_futuras))
print("Primera fecha:", fechas_futuras['Date'].min())
print("Última fecha:", fechas_futuras['Date'].max())
#Veamos las fechas predichas y que cumplan con lo requerido
```

```python
plt.figure(figsize=(12,6))

plt.plot(
    train_limpio['Date'],
    train_limpio['Weekly_Sales'],
    label="Histórico"
)

plt.plot(
    fechas_futuras,
    predicciones_real.flatten(),
    label="Forecast",
    linestyle="--"
)

plt.legend()
plt.show()
```
